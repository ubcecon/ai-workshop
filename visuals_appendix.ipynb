```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# Create an empty network graph
G = nx.Graph()

# Define nodes and edges
nodes = ("Paul", "John", "Helen", "Mirriam", "Luke", "Davis")
edges = [("Paul", "John"), ("John", "Helen"), ("Helen", "Luke"), ("Mirriam", "Luke"), 
         ("Davis", "Luke"), ("Mirriam", "John"), ("Mirriam", "Paul"), 
         ("Paul", "Helen"), ("Davis", "Paul"), ("Luke", "John"), ("Helen", "Mirriam")]

# Add nodes and edges to the graph
G.add_nodes_from(nodes)
G.add_edges_from(edges)

# Compute closeness centrality
centrality = nx.closeness_centrality(G)
centrality_values = np.array(list(centrality.values()))
centrality_standard = np.argsort(np.argsort(centrality_values))
norm = plt.Normalize(min(centrality_standard), max(centrality_standard), clip=False) 
cmap = plt.cm.YlGnBu

# Normalize centrality values for coloring
norm = plt.Normalize(vmin=min(centrality_values), vmax=max(centrality_values))
cmap = plt.cm.YlGnBu

# Compute node positions only once
pos = nx.random_layout(G, seed=100)

# Set up the plot
fig, ax = plt.subplots(figsize=(10, 6))
plt.subplots_adjust(left=0.02, right=0.98, top=0.98, bottom=0.02)

# Draw nodes, edges, and labels using the same positions
nx.draw_networkx_nodes(G, pos=pos, node_size=2500, 
                       node_color=centrality_standard, cmap=cmap, edgecolors="black", alpha=1, ax=ax)
nx.draw_networkx_edges(G, pos=pos, alpha=0.6, ax=ax)
nx.draw_networkx_labels(G, pos=pos, ax=ax)

# Create a ScalarMappable and add a color bar
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
plt.colorbar(sm, ax=ax)

# Save the plot
plt.savefig("example_graph_closeness_centrality.png")
plt.show()
```
